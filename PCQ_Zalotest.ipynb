# III. CLEANING DATA 

## 1. df_access_log

```python
df_access_log.head()
```

```python
# insert column user_funnel by 
df_access_log['user_funnel_access_log'] = df_access_log['url_path'].apply(lambda x: 'Landing on main page' if 'https://social.zalopay.vn/spa/telco/phonecard' in x else 'Non')
df_access_log_clean = df_access_log[['campaign_name','zaloid','from_menu','from_source','user_funnel_access_log']]
df_access_log_clean = df_access_log_clean[df_access_log_clean['user_funnel_access_log'] == 'Landing on main page']
df_access_log_clean.head()
```

```python
df_access_log_clean.shape
```

## 2. df_action_log

```python
df_action_log.head()
```

```python
df_action_log['user_funnel_action_log'] = 'pay attempt'
df_action_log['userid'] = df_action_log['zalopayid']
df_action_log_clean = df_action_log[['zaloid','zalopayid','userid','user_funnel_action_log']]
df_action_log_clean.head()
```

```python
df_action_log_clean.shape
```

## 3. df_tpe

```python
df_tpe.head()
```

```python
# Extract price of each card 
df_tpe['price_card'] = df_tpe['item'].apply(lambda x: (((x.split('Mệnh giá:')[1]).split('đ')[0]).replace('.','')))
df_tpe.price_card = df_tpe.price_card.astype(float)
# Extract promotion value of each card
df_tpe['promo_per_card'] = df_tpe['item'].apply(lambda x: (((x.split('Chiết khấu:')[1]).split('đ')[0]).replace('.','').replace('-','')))
df_tpe.promo_per_card = df_tpe.promo_per_card.astype(float)
# Extract number of card users bought
df_tpe['volumn_bought'] = df_tpe['item'].apply(lambda x: int(((x.split('Số lượng:')[1]).replace('"}',''))))
# Calculate total promotion amount 
df_tpe['total_promo_amount_tpe'] = df_tpe['promo_per_card'] * df_tpe['volumn_bought']
```

```python
df_tpe.head()
```

```python
df_tpe['user_funnel_tpe'] = df_tpe['item'].apply(lambda x: 'pay success' if 'Viettel' in x else 'Non')
df_tpe_clean = df_tpe[['transid','userid','amount','userchargeamount','total_promo_amount_tpe','volumn_bought', 'user_funnel_tpe']]
df_tpe_clean = df_tpe_clean.rename(columns = {'amount':'amount_tpe'})
df_tpe_clean.head()
```

```python
df_tpe_clean.shape
```

## 4. df_promotion_details

```python
df_promotion_details.head()
```

```python
df_promotion_details['user_funnel_promo'] = 'Promotion users'
df_promotion_details_clean = df_promotion_details[['transID', 'userID', 'amount', 'userCharge', 'new Map', 'new Pay', 'Số tiền được cashback', 'user_funnel_promo']]
df_promotion_details_clean = df_promotion_details_clean.rename(columns={'transID':'transid', 'amount':'amount_promotion'})

df_promotion_details_new_map = df_promotion_details_clean[df_promotion_details_clean['new Map'] == 1]
df_promotion_details_new_map_clean = df_promotion_details_new_map[['transid']]
df_promotion_details_new_map_clean['user_funnel_promo_map'] = 'New map users'

df_promotion_details_new_pay = df_promotion_details_clean[(df_promotion_details_clean['new Pay'] == 1)&(df_promotion_details_clean['new Map'] == 1)]
df_promotion_details_new_pay_clean = df_promotion_details_new_pay[['transid']]
df_promotion_details_new_pay_clean['user_funnel_promo_pay'] = 'New paying users'
```

```python
df_merge_action_tpe = pd.merge(df_merge_access_action, df_tpe_clean, how='left', on = ['userid'])
df_merge_action_tpe.head()
```

```python
# Check duplicate, No duplicates found 
duplicateRowsDF = df_merge_action_tpe[df_merge_action_tpe.duplicated()]
duplicateRowsDF.shape
```

```python
# Check duplicate, No duplicates found 
duplicateRowsDF = df_merge_tpe_promo[df_merge_tpe_promo.duplicated()]
duplicateRowsDF.shape
```

### 1.5. Left join promotion_details_new_map and promotion_details_new_pay
